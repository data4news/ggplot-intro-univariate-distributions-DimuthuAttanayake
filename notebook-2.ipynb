ðŸ‘‰ Bonus

```python
%%R -w 1000 -h 200

x = df$pct_below_poverty

q1  = quantile(x, 0.25, na.rm = TRUE)
q3  = quantile(x, 0.75, na.rm = TRUE)
iqr <- q3 - q1

outliers = df[
  x < q1 - 1.5 * iqr | x > q3 + 1.5 * iqr,
]

ggplot(df) + 
  aes(x = pct_below_poverty, y = 1) +
  geom_boxplot(alpha = 0.2, outlier.size = 10) +
  facet_wrap(~ borough)+
  geom_text(
    data = outliers,
    aes(x = pct_below_poverty, y = 1, label = zip),
    vjust = -1
  ) +
  labs(
    title = "Bronx is where the poorest peolpe in the New York City live",
    subtitle = "People living below poverty line in the New York City",
  )
```
